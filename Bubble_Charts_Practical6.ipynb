```python
import matplotlib.pyplot as plt
import numpy as np

# Data
x = np.random.rand(20)
y = np.random.rand(20)
z = np.random.rand(20)
colors = np.random.rand(20)
sizes = 1000 * np.random.rand(20)

# Plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z, c=colors, s=sizes, alpha=0.6)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 30, 40, 50]
y2 = [15, 25, 35, 45, 55]
sizes = [100, 200, 300, 400, 500]

# Plot
plt.scatter(x, y1, s=sizes, c='blue', alpha=0.5, label='Series 1')
plt.scatter(x, y2, s=sizes, c='red', alpha=0.5, label='Series 2')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.legend()
plt.show()
```
